```
# add contact
contact={ }
def addContact(name,phone):
  if name not in contact:
    contact[name]=phone
    print('contact details are added')
  else :
    print('contact already exists')
  return
addContact('anil','1234567890')
addContact('vijay','9874563210')
addContact('anil','1234567890')
print(contact)
```

```
def searchContact(name):
  if name in contact:
    print(name, ":",contact[name])
  else:
    print('%s does not exists'% name)
  return
searchContact('anil')
searchContact('vijay')
searchContact('deelip')
```

```
def importcontact(newContact):
  contact.update(newContact)
  print(len(newContact.keys()),"Contacts succesfully added")
  return
newContact={'diensh':1523641320,"Ajay":9977556632}
importcontact(newContact)
```

```
# delete a contact
def deleteContact(name):
  if name in contact:
    del contact[name]
    print(name,"deleted successfully")
  else:
    print(name,"not exists")
  return
deleteContact('anil')
deleteContact('dileep')
```
