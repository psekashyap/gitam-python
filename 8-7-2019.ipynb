```
#given number is prime number or not
def isPrime(n):
  flag = True
  for i in range(2,n//2+1) :
    if n%1==0:
      flag = False
      return flag
  return flag
isPrime(12)
```

```
# function to find prime numbers from 1 to 10
def primecount(n):
  cnt=0
  for a in range(2,n+1):
    k=0
    for i in range(2,a//2+1):
      if a%i==0:
        k=k+1
      if(k<=0):
        cnt=cnt+1
  return cnt
primecount(10)
```

```
# return the number of palindrome numbers
def palindrome(n,m):
  count=0
  for n in range (m):
    r=n%10
    n=n//10
    if r==n :
      count+=1
    n=n+1
  print(count)
  return
palindrome(11,99)
```

```
#fuction to get the perfect number
 def factorslist(n):
    sum=0
    for i in range(1,n):
      if n%i==0 :
        sum=sum+i
    if sum==n:
      print ("perfect number")
    else : 
      print("not perfect")
    return
 factorslist(6)
```
