```
def secondlarge(li):
  li.sort()
  return li[-2]
def genericlarge(li,n):
  li.sort()
  return li [-n]
def secondleast(li):
  li.sort()
  return li[1]
def genricleast(li,n):
  li.sort()
  return li[n]
li=[1,19,6,2,8,18,3]
genricleast(li,1)
```

```
#1 first and last element print as it is
#2 both sides numbers are even  then print it
li1=[2,5,6,7,4,3,8]
def printList(li1):
  for x in range(len(li1)):
    if x==0 or x==len(li1)-1:
      print(li1[x],end=" ")
    elif li1[x-1]%2==0 and li1[x+1]%2==0:
      print(li1[x],end=" ")
  return
printList(li1)
```

```
def numberlistconversion(n):
  li=[]
  while n!=0:
    r=n%10
    li.append(r)
    n=n//10
    li.reverse()
    return li
  numberlistconversion(1234)
```
