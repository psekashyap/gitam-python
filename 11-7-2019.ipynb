```
def readFile(filename):
  f=open(filename,'r')
  if f.mode==('r'):
    x=f.read()
    print(x)
  f.close()
  return
readFile('file2.txt')
```

```
#word count program
def wordCount(filename,word):
  with open(filename,'r')as f:
    if f.mode=='r':
      x=f.read()
      li=x.split()
  cnt=li.count(word)
  return cnt
filename=input('enter the filename:')
word=input('enter word:')
wordCount(filename,word)
```

```
def numberofLines(filename):
  lines=0
  with open(filename,'r') as f:
    if f.mode=='r':
      for line in f:
        lines+=1
  return lines
filename=input('enter the file name')
numberofLines(filename)
      
```

```
# count number of uppercase and lower case letters
def caseCount(filename):
  cntupper=0
  cntlower=0
  with open(filename,'r') as f:
    if f.mode=='r':
      x=f.read()
      li=list(x)
    for i in li:
      if i.isupper():
        cntupper+=1
      elif i.islower():
        cntlower+=1
    output='Upper Case={0},Lower Case={1}'.format(cntupper,cntlower)
  return output
filename=input('enter the name of file')
caseCount(filename)
```
